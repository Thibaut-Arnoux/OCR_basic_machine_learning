```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
house_data = pd.read_csv('house.csv')
house_data = house_data[house_data['loyer'] < 10_000]
```

```python
house_data
```

```python
plt.scatter(house_data['surface'], house_data['loyer'])
```

```python
X = np.array([house_data['surface']]).T
X
```

```python
y = np.array([house_data['loyer']]).T
y
```

```python
reg = linear_model.LinearRegression()
reg.fit(X, y)
reg.coef_.item(0)
plt.plot([0, 250], [reg.coef_.item(0), 250 * reg.coef_.item(0)], linestyle='--')
plt.scatter(house_data['surface'], house_data['loyer'])
plt.show()
```
